# Load Data


````python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/eshibhatt/Fall-Detection/main/Train.csv')
df

````

````python
df_test = pd.read_csv('https://raw.githubusercontent.com/eshibhatt/Fall-Detection/main/Test.csv')
````

## Data Preparation

### Data Seperation as X and y

````python
y_train = df['fall']
y_train
````

````python
X_train = df[['acc_max','gyro_kurtosis','lin_max' , 'acc_skewness' , 'gyro_skewness' , 'post_gyro_max' , 'post_lin_max']]
X_train
````

````python
y_test = df_test['fall']
y_test
````

````python
X_test = df_test[['acc_max','gyro_kurtosis','lin_max' , 'acc_skewness' , 'gyro_skewness' , 'post_gyro_max' , 'post_lin_max']]
X_test
````

# Model Building

````python
import numpy as np
````

````python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
````

````python
model = LogisticRegression()
````

````python
model.fit(X_train, y_train)
````



```
# This is formatted as code
```

# Model Testing


````python
y_pred = model.predict(X_test)
````

````python
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_pred)
````

````python
print(f"Accuracy: {accuracy * 100:.2f}%")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1-Score: {f1:.2f}")
print(f"ROC AUC: {roc_auc:.2f}")
````
