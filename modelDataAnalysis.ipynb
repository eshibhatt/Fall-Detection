```python
import pandas as pd
from pandas_profiling import ProfileReport
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/eshibhatt/Fall-Detection/main/Train.csv')
df
```

```python
unnamed_column_index = 0
```

```python
df_final = df.drop(df.columns[unnamed_column_index], axis=1)
df_final
```

```python

profile = ProfileReport(df_final)
```

```python
# Save the report to an HTML file
profile.to_file("report.html")
```

```python
from google.colab import files
files.download('report.html')
```

```python
correlation_matrix = df[['acc_max', 'gyro_kurtosis', 'lin_max', 'acc_skewness', 'gyro_skewness', 'post_gyro_max', 'post_lin_max', 'fall']].corr()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 6))
```

```python
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
print(correlation_matrix['fall'].sort_values(ascending = False),'\n')
```
